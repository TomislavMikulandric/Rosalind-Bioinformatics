```python
def reconstruct_from_genome_path(kmers):
    if not kmers:
        return ""
    genome = kmers[0]
    k = len(kmers[0])

    for i in range(1, len(kmers)):
        genome += kmers[i][-1]

    return genome

if __name__ == "__main__":
    with open("../data/rosalind_ba3b.txt") as f:
        kmers = [line.strip() for line in f]
    result = reconstruct_from_genome_path(kmers)
    print(result)
```
