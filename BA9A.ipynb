```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

def construct_trie(patterns):
    root = TrieNode()
    node_counter = 1
    node_map = {root: 0}
    edges = []

    for pattern in patterns:
        current_node = root
        for char in pattern:
            if char not in current_node.children:
                new_node = TrieNode()
                current_node.children[char] = new_node
                node_map[new_node] = node_counter
                edges.append((node_map[current_node], node_counter, char))
                node_counter += 1
            current_node = current_node.children[char]
        current_node.is_end = True

    return edges

def main():
    with open("../data/rosalind_ba9a.txt") as f:
        patterns = [line.strip() for line in f]
    trie_edges = construct_trie(patterns)

    for edge in trie_edges:
        print(f"{edge[0]}->{edge[1]}:{edge[2]}")

if __name__ == "__main__":
    main()
```
