```python
from itertools import product

def generate_kmer(k):
    return [''.join(p) for p in product("ACGT", repeat=k)]

def compute_frequency_array(text, k):
    # Generate all possible k-mers
    all_kmers = [''.join(p) for p in product('ACGT', repeat=k)]
    
    # Initialize frequency array with zeros
    freq_array = [0] * (4**k)
    
    # Count occurrences of each k-mer in the text
    for i in range(len(text) - k + 1):
        kmer = text[i:i+k]
        index = all_kmers.index(kmer)
        freq_array[index] += 1
    
    return freq_array


if __name__ == "__main__":
    with open("../data/rosalind_ba1k.txt") as f:
        text = f.readline().strip()
        k = int(f.readline().strip())
    result = compute_frequency_array(text, k)
    print(' '.join(map(str, result)))
```
