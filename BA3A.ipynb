```python
def string_composition(k, text):
    composition = []
    for i in range(len(text)-k+1):
        kmer = text[i:i+k]
        composition.append(kmer)
    return sorted(composition)


if __name__ == "__main__":
    with open("../data/rosalind_ba3a.txt") as f:
        k = int(f.readline().strip())
        dna = f.readline().strip()
    result = string_composition(k, dna)
    for kmer in result:
        print(kmer)
```
