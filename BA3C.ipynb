```python
def construct_overlap_graph(kmers):
    adjacency_list = []
    for i, kmer1 in enumerate(kmers):
        for j, kmer2 in enumerate(kmers):
            if i != j and kmer1[1:] == kmer2[:-1]:
                adjacency_list.append(f"{kmer1} -> {kmer2}")
    return sorted(adjacency_list)


if __name__ == "__main__":
    with open("../data/rosalind_ba3c.txt") as f:
        patterns = [line.strip() for line in f]
    edges = construct_overlap_graph(patterns)
    for edge in edges:
        print(edge)
```
